```python
import os
import sympy as sp
import numpy as np
from multipie.core.model_analyzer import ModelAnalyzer

MODEL_DIR = os.getcwd() + "/model/"
graphene = {
    "samb": {
        "model": "graphene",
        # "parameter": {
        #     "z4": 0.0, # chiral CEF
        #     "z6":1.0, # nonchiral hopping
        #     "z8":0.2, # chiral hopping
        # },
        "samb_figure": False,
        "k_multipole":True,
    },
    "wannier": {  
        "seedname": "graphene",  # read seedname.win, seedname.nnkp, (seedname.mmn, seedname.spn, seedname.uHu, seedname.uIu ...).
        # "ket_wannier": [
        #     ['A', 2, 1, 2, 'pz'],
        #     ['A', 1, 1, 2, 'pz'],
        # ],
        "hr_file": "graphene_hr_cw.dat",  # read seedname_hr.dat.
    },
    "output": {
        "dispersion": {
            "k_path":"",
            "local":["Lz"],
        },
    },
}

ma = ModelAnalyzer(N1=200,topdir=MODEL_DIR, verbose=True)
ma.analyze(graphene)
```
