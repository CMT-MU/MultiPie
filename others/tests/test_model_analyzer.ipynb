```python
from multipie.util.util import str_to_sympy
# from multipie.util.util_model_analyzer import fourier_r_to_k

def fourier_r_to_k(OR, atom, kv, s=True):
    d = len(atom)
    S = np.full((d, d), sp.S(0), dtype=object)
    for (n1, n2, n3, m, n), value in OR.items():
        if s:
            Rs = np.array([n1,n2,n3]) - atom[m] + atom[n]
        else:
            Rs = np.array([n1,n2,n3])
        S[m,n] += value * sp.exp(sp.I*(kv @ Rs))

    return S



tb_gauge = ma.output["fourier"]["tb_gauge"]
n = len(ma.model.get_ket_site())
atom = np.asarray(sp.symbols(' '.join(f'kb_{i}' for i in range(1, len(ma.model.get_ket_site())+1))),dtype=object)
atom = np.array(sp.symbols(" ".join([f"r_{i}^(1) r_{i}^(2) r_{i}^(3)" for i in range(1,n+1)])), dtype=object).reshape(-1,3)
# np.asarray(list(ma.model.get_ket_site().values()), dtype=float)
tag = "z9"
d = ma.md["matrix"]["z9"]
d = {k:str_to_sympy(v) for k, v in d.items()}
kv = str_to_sympy("[k1,k2,k3]")
mat = fourier_r_to_k(d, atom, kv)
sp.Matrix(mat)
```
