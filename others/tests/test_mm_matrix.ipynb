```python
import os
import sympy as sp
from multipie import MaterialModel

MODEL_DIR = os.getcwd() + "/others/tests/model/"

mm = MaterialModel(topdir=MODEL_DIR, verbose=False)
```

```python
model = "graphene"
mm.load(model)
select = {"l": 0, "bond": [[0, 1, 2]]}
# select = {}
cid, sel, _ = mm.select_combined_samb(**select)
print("select:", select)
print("select SAMB:", sel)
for zi, (tag, info, index, comp) in cid.items():
    print(zi, (tag, info, index, comp))
combined_samb_matrix = mm.get_combined_samb_matrix(**select)
print("selected:", list(combined_samb_matrix.keys()))
```

```python
coeff = {"z1": sp.S(1), "z2": sp.S(1)}

print("-- combined SAMB matrix")
for zj in coeff.keys():
    d = combined_samb_matrix[zj]
    tag, info, index, comp = mm["combined_id"][zj]
    print(f"== {zj}, {info} ==")
    display(sp.Symbol(tag))
    for (n1, n2, n3, m, n), v in d.items():
        display(sp.Matrix([n1,n2,n3,m,n,v]).T)

print("-- Hamiltonian matrix")
H = mm.get_hr(coeff, combined_samb_matrix)
for (n1, n2, n3, m, n), v in H.items():
    display(sp.Matrix([n1,n2,n3,m,n,v]).T)
```
