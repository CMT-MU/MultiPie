```python
import sympy as sp
import numpy as np
from multipie import Group
```

```python
g = Group("D3")
d = g.response_tensor_all("Q")
for t, (m,ex) in d.items():
    print(t)
    if not np.all(m == sp.S(0)):
        display(sp.Matrix(m))
        for i,j in ex.items():
            display(sp.Eq(i,j))
```
