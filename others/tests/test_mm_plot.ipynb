```python
import os
from multipie import MaterialModel
from qtdraw import QtDraw

MODEL_DIR = os.getcwd() + "/others/tests/model/"

mm = MaterialModel(topdir=MODEL_DIR, verbose=True)
qtdraw = QtDraw()
```

```python
model = "Mn3Sn"
view = MODEL_DIR + model + "/" + model + ".qtdw"
mm.load(model)
group = mm.group
qtdraw.clear_data()
qtdraw.set_crystal(group.info.crystal)
if mm["group"].count("^")==0:
    qtdraw.set_cell("off")
else:
    qtdraw.set_cell("single")
```

```python
print(list(mm["wyckoff"].keys()))
site = "Sn"
qtdraw.clear_data()
mm.plot_site_bond(qtdraw,site)
bond = "Sn;Mn_001_1"
mm.plot_site_bond(qtdraw,bond)
```

```python
bond = "Sn;Mn_001_1"
qtdraw.clear_data()
mm.plot_cluster_samb(qtdraw, bond, "y52")
mm.plot_cluster_samb(qtdraw, bond, "y55")
```

```python
site = "Mn"
qtdraw.clear_data()
mm.plot_cluster_samb(qtdraw, site, "y4")
```
