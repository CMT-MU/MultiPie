```python
"""
Check for root-cluster harmonics binary.
"""
from multipie.util.util_binary import BinaryManager
from qtdraw import QtDraw
from multipie.core.multipie_info import __top_dir__
BIN_DIR = __top_dir__ + "others/tools/binary_data"
from utils.util_plot import _plot_scalar_harmonics, _plot_vector_harmonics
qtdraw = QtDraw()
qtdraw.set_cell("off")
info = BinaryManager("info", verbose=True, topdir=BIN_DIR)
rc = BinaryManager("harmonics_root_cluster", verbose=True, topdir=BIN_DIR)
rc_info = BinaryManager("root_cluster", verbose=True, topdir=BIN_DIR)
point = rc_info["position"]
```

```python
tag = "D6h"
no = info["id"][tag]
for idx, (b, ex) in rc[no]["monopole"].items():
    print(idx, ex)
```

```python
for idx, (b, ex, v_ex) in rc[no]["dipole"].items():
    print(idx, ex)
```

```python
X, l, Gamma, n, comp = 'Q', 1, 'A2u', -1, 0
idx = (X, l, Gamma, n, -1, 0, 0, 'q')
qtdraw.clear_data()
_plot_scalar_harmonics(qtdraw,point.astype(float),rc[no]["monopole"][idx][0][comp].astype(float))
```

```python
X, l, Gamma, n, k, comp = 'G', 1, 'A2g', -1, 0, 0
idx = (X, l, Gamma, n, -1, 1, k, 'q')
qtdraw.clear_data()
_plot_vector_harmonics(qtdraw,point.astype(float),rc[no]["dipole"][idx][0][comp].astype(float))
```
