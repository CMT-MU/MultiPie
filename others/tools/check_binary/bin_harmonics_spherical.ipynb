```python
"""
Check for spherical harmonics binary.
"""
import sympy as sp
from multipie.util.util_binary import BinaryManager
from qtdraw import QtDraw
from multipie.core.multipie_info import __top_dir__
BIN_DIR = __top_dir__ + "others/tools/binary_data"
from utils.util_plot import plot_harmonics
info = BinaryManager("info", verbose=True, topdir=BIN_DIR)
sh = BinaryManager("harmonics_spherical", verbose=True, topdir=BIN_DIR)
qtdraw = QtDraw()
qtdraw.set_cell("off")
```

```python
sh.keys()
```

```python
s, l, k = 2, 1, 1 # m = 1, 0, -1.
X = "Q" if (s+k)%2==0 else "G"
sp.Matrix(sh["harmonics"][(X,l,s,k,"q")] @ info["harmonics"]["internal_basis"][(s,"q")])
```

```python
rv = info["harmonics"]["variable"]
sv = info["harmonics"]["internal_basis"][(1,"q")]
sp.Matrix([info["harmonics"]["basis_function"][i][0].subs({rv[0]:sv[0],rv[1]:sv[1],rv[2]:sv[2]}) for i in map(lambda i:str(i).replace("_","").replace("Q","d"),info["harmonics"]["internal_basis"][(s,"q")])]).T
```

```python
v = info["harmonics"]["basis_function"]["f3"][1]
plot_harmonics(qtdraw, 3, v=v) # l=3, s=k=0, s32 only.
```

```python
v = info["harmonics"]["basis_function"]["pz"][1]
plot_harmonics(qtdraw, 1, 1, 0, v=v) # l=1, s=1, k=0, c10 only.
```
