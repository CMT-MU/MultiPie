```python
"""
Check for multipole harmonics binary.
"""
from multipie.util.util_binary import BinaryManager
from qtdraw import QtDraw
from multipie.core.multipie_info import __top_dir__
BIN_DIR = __top_dir__ + "others/tools/binary_data"
from utils.util_plot import plot_multipole_harmonics
info = BinaryManager("info", verbose=True, topdir=BIN_DIR)
harmonics = BinaryManager("harmonics_multipole", verbose=True, topdir=BIN_DIR) # it takes about 5 min.
qtdraw = QtDraw()
qtdraw.set_cell("off")
```

```python
harmonics.keys()
```

```python
tag, X, l = "D3", "G", 2
internal_type, s = "g", 3
no = info["id"][tag]
for idx, (ex, vb, v) in harmonics[no].select(s=s,x=internal_type,X=X,l=l).items():
    print(tuple(idx), ex.tolist(), vb.tolist(), v.tolist())
```

```python
for i in harmonics[no][('G', 2, 'E', 1, -1, s, 1, internal_type)][2]:
    display(i)
```
