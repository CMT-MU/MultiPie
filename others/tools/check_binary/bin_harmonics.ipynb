```python
"""
Check for harmonics binary.
"""
import sympy as sp
from multipie.util.util_binary import BinaryManager
from qtdraw import QtDraw
from multipie.core.multipie_info import __top_dir__
BIN_DIR = __top_dir__ + "others/tools/binary_data"
from utils.util_plot import plot_multipole_harmonics
info = BinaryManager("info", verbose=True, topdir=BIN_DIR)
harmonics = BinaryManager("harmonics", verbose=True, topdir=BIN_DIR)
qtdraw = QtDraw()
qtdraw.set_cell("off")
```

```python
harmonics.keys()
```

```python
tag, X, l = "D3", "G", 2
no = info["id"][tag]
harm = harmonics[no]["harmonics"].select(X=X,l=l)
print(*harm.field)
for idx, (c, U, t) in harm.items():
    print(tuple(idx))
    display(sp.Matrix(t).T)
    display(sp.Matrix(c).T)
    display(sp.Matrix(U))
```

```python
harmonics_pg = harmonics[no]
idx = ('G', 2, 'A1', -1, -1, 0, 0, 'q')
comp = -1
plot_multipole_harmonics(qtdraw,harmonics_pg,idx,comp)
```
