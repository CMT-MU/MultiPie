```python
"""
Check for atomic_multipole binary.
"""
import sympy as sp
from multipie.util.util_binary import BinaryManager
from multipie.core.multipie_info import __top_dir__
BIN_DIR = __top_dir__ + "others/tools/binary_data"
atomic_multipole = BinaryManager("atomic_multipole", verbose=True, topdir=BIN_DIR)
```

```python
bra, ket = 0, 1
for (X, l, s, k, x), mat in atomic_multipole["lm"][(bra,ket)].items():
    print(f"({X},{l},{s},{k})", ", m =", list(range(l,-l-1,-1)))
    for m in mat:
        display(sp.Matrix(m))
```
