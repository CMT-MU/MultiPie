```python
"""
Check for atomic_multipole binary.
"""
import numpy as np
import sympy as sp
from multipie.util.util_binary import BinaryManager
from multipie.core.multipie_info import __top_dir__
BIN_DIR = __top_dir__ + "others/tools/binary_data"
info = BinaryManager("info", verbose=True, topdir=BIN_DIR)
atomic_multipole = BinaryManager("atomic_multipole_group", verbose=True, topdir=BIN_DIR)
```

```python
atomic_multipole.keys()
```

```python
tag = "Oh"
no = info["id"][tag]
am = atomic_multipole[no]["lg"][(1,1)]
print(*am.field)
for idx, (v, ex) in am.select(s=0).items():
    print(tuple(idx))
    for e, m in zip(ex,v):
        display(e)
        m = sp.Matrix(m)
        display(m)
        display(np.vdot(m,m))
```
