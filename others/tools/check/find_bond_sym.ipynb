```python
from multipie import Group

so_dic = {}
for id_s in Group.global_info()["id_set"]["PG"]["all"]:
    group = Group(id_s)
    so_dic[str(sorted(group["symmetry_operation"]["tag"]))] = id_s

for id_s in Group.global_info()["id_set"]["PG"]["all"]:
    group = Group(id_s)
    wyckoff_bond = group["wyckoff"]["bond"]
    so_tag = group["symmetry_operation"]["tag"]
    for wb, val in wyckoff_bond.items():
        mp = val["mapping"][0]
        mp = str(sorted([so_tag[abs(i)-1] for i in mp]))
        if mp not in so_dic.keys():
            mp = mp.replace("2[100]", "2[010]")
            mp = mp.replace("2[001]", "2[010]")
            if mp not in so_dic.keys():
                print("cannot find", id_s, wb, mp)
                continue
        # print(wb, mp, so_dic[mp])
```
